```python
i = 0
j = 0
map_num = 20
total_try = 200
while j < total_try and i < map_num:
    cfg = clpg.function_call.CityFunctionCall(seed=j+400)
    num_roads = 10
    success = cfg.generate_road_network(num_roads)
    if not success:
        j += 1
        continue
    cfg.generate_building_alone_roads()
    # cfg.generate_detail_around_buildings()
    cfg.generate_detail_alone_roads()
    cfg.export_city(f"output/single_world/total/demo_map/big_map_{num_roads}_{i}")
    main(f"output/single_world/total/demo_map/big_map_{num_roads}_{i}", True)
    j += 1
    i += 1
```
