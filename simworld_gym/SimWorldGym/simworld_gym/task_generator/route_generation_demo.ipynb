```python
def robot_perform_action(action):
    robot_name = 'spot_robot'
    speed = 5000
    duration = 0.1
    if action == 0:
        route_generator.unrealcv_client.apply_action_transition(robot_name, [speed, duration, 0])
    elif action == 1:
        route_generator.unrealcv_client.apply_action_transition(robot_name, [speed, duration, 1])
    elif action == 2:
        route_generator.unrealcv_client.apply_action_transition(robot_name, [speed, duration, 2])
    elif action == 3:
        route_generator.unrealcv_client.apply_action_transition(robot_name, [speed, duration, 3])
    elif action == 4:
        route_generator.unrealcv_client.apply_action_rotation(robot_name, [duration, 90, 1])
    elif action == 5:
        route_generator.unrealcv_client.apply_action_rotation(robot_name, [duration, -90, -1])
    time.sleep(duration+0.05)
    # Wait until the action in Unreal Engine has finished executing
    is_available = route_generator.unrealcv_client.get_is_available(robot_name)
    while not is_available:
        is_available = route_generator.unrealcv_client.get_is_available(robot_name)
        time.sleep(0.05)
```

```python
def wirte_current_step_data(image, current_step_path, current_instruction, current_visual_image, past_actions, ground_truth_action):
    cv2.imwrite(os.path.join(current_step_path, f"current_observation.png"), image)
    cv2.imwrite(os.path.join(current_step_path, f"visual_instruction.png"), current_visual_image)
    data = {
                "text_instruction": current_instruction,
                "visual_instruction": f"visual_instruction.png",
                "past_actions": past_actions,
                "current_observation": "current_observation.png",
                "ground_truth_action": ground_truth_action
            }
    with open(os.path.join(current_step_path, f"meta_data.json"), "w", encoding="utf-8") as f:
        json.dump(data, f, indent=4, ensure_ascii=False)
```
